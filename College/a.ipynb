```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,20]
plt.plot(x,y)
plt.title("graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y,color='green',linestyle='-.')
```

```python
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart")
plt.show()
```

```python
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [30, 25, 25, 20]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart")
plt.axis('equal')
plt.show()
```
